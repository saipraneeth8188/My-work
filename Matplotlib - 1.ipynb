```python
x=[1,2,3]
store1=[10,20,5]
store2=[12,14,3]
plt.bar(x,store1,width=0.2,label='store1')
plt.bar(x,store2,width=0.2,label='store2')
plt.legend()
plt.show()
```

In the above graph you can see in the first bar , the blue bar is hiding under orange , refer below code for better understanding

```python
x=[1,2,3]
store1=[10,20,5]
store2=[12,14,3]
plt.bar(x,store1,width=0.2,label='store1',alpha=0.2)
plt.bar(x,store2,width=0.2,label='store2',alpha=0.2) # alpha is like brightness
plt.legend()
plt.show()
```

So you can see value of blue is 10 and orange is 12 , so blue is hiding under orange , what is the solution to overcome this

```python
plt.bar(x,store1,width=0.2,label='store1')
plt.bar([1.2,2.2,3.2],store2,width=0.2,label='store2') 
plt.legend()
plt.show()

# in the second plot we are plotting it in 1.2,2.2 and 3.2 so we could see a difference
```

```python
x=np.arange(1,4)
plt.bar(x,store1,width=0.2,label='store1')
plt.bar(x+0.2,store2,width=0.2,label='store2') 
plt.legend()
plt.show()
```

# SUBPLOTS

```python
plt.figure(figsize=(5,6))
plt.plot(x,y)
plt.show()

# Figsize is used to modify the size of plots
```

```python
x=np.arange(1,10,0.1)
y=np.sin(x)
y2=np.tan(x)
y3=np.log(x)

plt.figure()

plt.subplot(2,3,1)  # it creates 2 rows and 3 columns and put this image in 1st chamber
plt.plot(x,y,color='red')

plt.subplot(2,3,3)  # it puts this plot in the 3rd chamber
plt.plot(x,y2,color='green')

plt.subplot(2,3,5) # it puts the plot in the 5th chamber
plt.plot(x,y3,color='yellow')

plt.show()
```

```python
fig=plt.figure() # it is nothing but a blank paper
```

```python
fig=plt.figure()

ax231=fig.add_subplot(2,3,1)
ax231.plot(x,y)

ax233=fig.add_subplot(2,3,3)
ax233.plot(x,y2)
ax233.set_xlim(-5,10)

ax235=fig.add_subplot(2,3,5)
ax235.plot(x,y3)
ax235.set_ylim(-10,10)

plt.show()
```
